### Tener en cuenta en multiplicación y potenciación

Es importante tener en cuenta que la multiplicación matricial y la elevación de matrices no son las mismas operaciones y, por lo tanto, no producen los mismos resultados. Aunque la multiplicación matricial se puede utilizar para calcular la potencia de una matriz, no es lo mismo que elevar una matriz a una potencia determinada.

Cuando multiplicamos una matriz A por sí misma, el resultado es otra matriz B, y si multiplicamos B por A de nuevo, obtenemos otra matriz C. Cada una de estas matrices tiene una estructura diferente, ya que cada una representa una transformación lineal compuesta por la transformación representada por la matriz original A. Por lo tanto, la multiplicación matricial 3 veces por sí misma no produce el mismo resultado que elevar la matriz al cubo.

Por otro lado, la elevación de una matriz a una potencia determinada se define como la multiplicación de la matriz por sí misma un número específico de veces. En otras palabras, la potencia k-ésima de una matriz A se define como:

A^k = A x A x A x ... x A (k veces)

Cada una de las multiplicaciones se realiza utilizando la misma matriz A, lo que significa que cada una de las transformaciones lineales es la misma que la representada por A. Por lo tanto, elevar una matriz al cubo produce un resultado diferente a la multiplicación matricial 3 veces por sí misma, ya que cada operación representa una transformación lineal diferente.

### Suma de Matrices

se utiliza para combinar matrices y realizar transformaciones lineales. La suma de dos matrices A y B del mismo tamaño se define como la matriz C que tiene los mismos tamaños que A y B, y cuyos elementos se calculan como la suma de los elementos correspondientes de A y B:

Cij = Aij + Bij

donde i es el índice de la fila y j es el índice de la columna.

image.png

Es importante tener en cuenta que la suma de matrices solo se puede realizar si las dos matrices tienen el mismo tamaño. Además, la suma de matrices es una operación conmutativa, es decir, A + B = B + A, y es una operación asociativa, es decir, (A + B) + C = A + (B + C).

En particular, la suma de matrices es importante en el análisis de redes, donde las matrices se utilizan para representar las conexiones entre nodos. En este caso, la suma de matrices se utiliza para combinar las conexiones entre diferentes nodos en una red y analizar la estructura y las propiedades de la red.

### Resta de Matrices
La resta de dos matrices A y B del mismo tamaño se define como la matriz C que tiene los mismos tamaños que A y B, y cuyos elementos se calculan como la resta de los elementos correspondientes de A y B:

Cij = Aij - Bij

donde i es el índice de la fila y j es el índice de la columna.

image.png

Es importante tener en cuenta que la resta de matrices solo se puede realizar si las dos matrices tienen el mismo tamaño. Además, la resta de matrices no es una operación conmutativa, es decir, A - B no es lo mismo que B - A.

La resta de matrices también se utiliza en la eliminación de ruido en imágenes y señales, donde se puede restar una matriz que representa el ruido de una matriz que representa la señal original para obtener una matriz que representa la señal sin ruido.

## Inversión y determinantes de matrices

 se utilizan para analizar y resolver sistemas de ecuaciones lineales y transformaciones lineales.

### Inversión
 Se refiere a la capacidad de deshacer una transformación lineal. 
 
 En otras palabras, si tenemos una matriz A que representa una transformación lineal, la inversa de A es una matriz A^-1 tal que
 
  A x A^-1 = A^-1 x A = I,
 
  
donde I es la matriz identidad. En términos prácticos, la inversa de una matriz se utiliza para resolver sistemas de ecuaciones lineales, ya que permite deshacer una transformación lineal y obtener los valores originales de las variables.

Sin embargo, no todas las matrices tienen una inversa. Una matriz es invertible o no singular si y solo si su determinante es diferente de cero.

Calcular la inversa de una matriz es sencillo:

````python
Ainv = np.linalg.inv(A)
Ainv
````

La matriz anterior puede no parecerse a la matriz identidad, pero si miramos más de cerca, veremos que las diagonales son todas 1 y los elementos fuera de la diagonal son un número muy pequeño (que desde el punto de vista de una computadora es 0).


### El determinante 

Es una función que asigna a cada matriz cuadrada un número escalar único que proporciona información sobre la transformación lineal que representa la matriz. El determinante de una matriz se denota como |A| y se calcula como una suma de productos de elementos de la matriz. El determinante se utiliza para determinar si una matriz es invertible o no singular. Una matriz tiene una inversa si y solo si su determinante es diferente de cero. Además, el determinante se utiliza para calcular la matriz adjunta, que se utiliza para encontrar la inversa de una matriz.

Para calcular el determinante

````python
np.linalg.det(A)
````

## Procesamiento de imágenes: 


la compresión de imágenes utilizando la descomposición en valores singulares (SVD). La descomposición en valores singulares es una técnica de álgebra lineal que se utiliza para descomponer una matriz en tres matrices: una matriz de valores singulares, una matriz de vectores singulares izquierdos y una matriz de vectores singulares derechos. La descomposición SVD se puede utilizar para reducir la complejidad de una imagen sin perder demasiada información.

Supongamos que tenemos una imagen en escala de grises representada como una matriz A de tamaño m x n. Podemos aplicar la descomposición SVD a la matriz A para obtener tres matrices: U, S y V, donde U es una matriz de vectores singulares izquierdos, S es una matriz diagonal de valores singulares y V es una matriz de vectores singulares derechos. Podemos aproximar la matriz A utilizando las primeras k columnas de U, las primeras k filas y columnas de S, y las primeras k filas de V. Esta aproximación se puede representar como:

A_k = U[:,:k] @ S[:k,:k] @ V[:k,:]

donde k es el número de valores singulares utilizados para la aproximación.

El siguiente código muestra cómo comprimir una imagen utilizando la descomposición SVD en Python:

````python
import numpy as np
from PIL import Image

# Cargar la imagen en escala de grises
img = Image.open('/content/image.jpg').convert('L')
A = np.array(img)
img.show(A)
````

````python

# Realizar la descomposición SVD
U, S, V = np.linalg.svd(A)

# Definir el número de valores singulares utilizados para la aproximación
k = 100

# Aproximar la matriz A utilizando las primeras k columnas de U, las primeras k filas y columnas de S, y las primeras k filas de V
A_k = U[:,:k] @ np.diag(S[:k]) @ V[:k,:]

# Convertir la matriz aproximada en una imagen
img_k = Image.fromarray(np.uint8(A_k))

# Mostrar la imagen original y la imagen aproximada
img.show()
img_k.show()
````

En este ejemplo, se carga una imagen en escala de grises utilizando la biblioteca PIL. A continuación, se realiza la descomposición SVD de la matriz de la imagen utilizando la función "np.linalg.svd" de NumPy. Se define el número de valores singulares utilizados para la aproximación y se aproxima la matriz de la imagen utilizando las primeras k columnas de U, las primeras k filas y columnas de S, y las primeras k filas de V. La matriz aproximada se convierte en una imagen utilizando la función "Image.fromarray" de PIL y se muestra la imagen original y la imagen aproximada en la salida.

## Optimización: 

La optimización de una cartera de inversión se refiere a la selección de una combinación óptima de activos que maximice el rendimiento esperado de la cartera, dado un nivel de riesgo deseado.

Supongamos que tenemos una cartera de inversión compuesta por tres activos: acciones, bonos y efectivo. Cada activo tiene un rendimiento esperado y una desviación estándar asociados, y queremos encontrar la combinación óptima de los tres activos que maximice el rendimiento esperado de la cartera, dado un nivel de riesgo máximo.

Para resolver este problema, podemos utilizar la teoría de la cartera de Markowitz, que utiliza la optimización cuadrática para encontrar la combinación óptima de activos. La solución del problema se puede expresar como un sistema de ecuaciones lineales utilizando las siguientes ecuaciones:


```
r = w^T mu
sigma^2 = w^T Sigma w
w^T 1 = 1
```

donde r es el rendimiento esperado de la cartera, mu es el vector de rendimientos esperados de los activos, Sigma es la matriz de covarianza de los activos, w es el vector de ponderaciones de los activos en la cartera, sigma^2 es la varianza de la cartera, y 1 es un vector de unos.

El objetivo es maximizar el rendimiento esperado de la cartera r, dado un nivel de riesgo máximo sigma_max, sujeto a la restricción de que la suma de las ponderaciones de los activos sea igual a uno.


````python
import numpy as np
import cvxpy as cp

# Definir los rendimientos esperados de los activos y la matriz de covarianza
mu = np.array([0.1, 0.05, 0.03])
Sigma = np.array([[0.1, 0.04, 0], [0.04, 0.2, 0], [0, 0, 0.05]])

# Definir el nivel de riesgo máximo
sigma_max = 0.1

# Definir las variables de optimización
w = cp.Variable(3)

# Definir la función objetivo y las restricciones
objective = cp.Maximize(mu @ w)
constraints = [cp.quad_form(w, Sigma) <= sigma_max**2, sum(w) == 1]

# Resolver el problema de optimización
prob = cp.Problem(objective, constraints)
prob.solve()

# Obtener la solución
w_opt = w.value

print('Las ponderaciones óptimas son:', w_opt)
````

En este ejemplo, se define el vector de rendimientos esperados de los activos y la matriz de covarianza. A continuación, se define el nivel de riesgo máximo. Se utilizan las funciones "cp.Variable" y "cp.quad_form" de CVXPY para definir las variables de optimización y la función objetivo. Se utilizan las restricciones definidas anteriormente para definir las restricciones del problema de optimización. Finalmente, se resuelve el problema de optimización utilizando la función "prob.solve()" de CVXPY y se muestra la solución
